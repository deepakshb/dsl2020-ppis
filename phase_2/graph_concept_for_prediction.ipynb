```
import numpy as np
import requests 
import networkx as nx
import pylab as plt
import matplotlib.pyplot as plt
from itertools import combinations
%matplotlib inline
```

```
data = requests.get('http://snap.stanford.edu/deepnetbio-ismb/ipynb/yeast.edgelist')

with open('yeast.edgelist', 'w') as file_edgelist:
    file_edgelist.write(data.text)
G = nx.read_edgelist('yeast.edgelist')
```

```
print(len(G.nodes()))
```

**Imporatce of the node** <br>
To see that how important the node is for in the given graph

There are two kind of centrality 


1.   Degree Centrality 
2.   Betweeness centrality

Here, In our case the degree centrality will be useful to know as it shows 

Let' see the number of neighbors for all the nodes


```
# Let us see the number of neighbor eace node have
# We are creating a dictionary for node and neighbors as key and value respectively 
degree = {n : len(list(G.neighbors(n))) for n in G.nodes()}
print(degree)
```

**Degree centrality** <br>
*degree centrality of the node is defined as the number of number of the neighbors node have to the number of neighbors node could possibly have* <br>

So, let's check the degre centrality for all the nodes in graph G 

```
# Lets create the list of the degree centrality for all the nodes
# Here degree is generated for the node in sequence from first to last node
degree_centrality_of_G = nx.degree_centrality(G)
print(degree_centrality_of_G)
```

**Connected component graph (CCG):** <BR>

*A set of nodes connected to one another by some path in the subgraph and note connected to other node in the larger graph*

below two graph are called connected component graphs 

```
plt.figure(1, figsize=(16, 12))
pos = nx.spring_layout(SG)

nx.draw_networkx_nodes(SG, pos, node_color='g', node_size=700)
nx.draw_networkx_edges(SG, pos)
nx.draw_networkx_labels(SG, pos, font_size=11)
plt.show()
```

```
# Betweenness centrality for the nodes 
betweenness_centrality_for_sg = nx.betweenness_centrality(SG)
print(betweenness_centrality_for_sg)
```

one of the simple complex clique we can consider is trianagle in the graph 

Let's check such triangle for the node 

```
total_triangle  = nx.triangles(G)
print(total_triangle)
```

So as above we can see it is returnig the dictioary where node is a key and value as a numbe of triagle to that node 

We can also find the clique for the graph using the the number of maximal clique for the graph which allows us to know the how many nodes are tightly connected in to the graph 

```
# Find the maximal clique in the graph
maximal_clique = nx.find_cliques(SG)
print(list(maximal_clique))
```
